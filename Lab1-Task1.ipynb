#Q3

```python
# Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
```

```python
# Compute the corresponding z values
z = x**2  + y**2
```

```python
# Create a figure and a 3D Axes.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Add labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Show the plot
plt.show()
```

```python
#import the seaborn and pandas library.

# seaborn -> library based on matplotlib that provides a high-level interface.
# pandas -> for data manipulation and analysis.

import seaborn as sns
import pandas as pd
```

```python
#download the data
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
```

```python
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
# Select the relevant features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]
```

```python
# Compute the Pearson correlation matrix
pearson_corr = data.corr(method='pearson')
```

```python
# Compute the Spearman rank correlation matrix
spearman_corr = data.corr(method='spearman')
```

```python
# Function to plot heatmap with values
def plot_heatmap(corr, title):
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
    plt.title(title)
    plt.show()
```

```python
# Plot the Pearson correlation heatmap
plot_heatmap(pearson_corr, 'Pearson Correlation Heatmap')
```

```python
# Plot the Spearman rank correlation heatmap
plot_heatmap(spearman_corr, 'Spearman Rank Correlation Heatmap')
```
